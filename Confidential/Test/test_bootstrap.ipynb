```python
import numpy as np
from sklearn.datasets import make_classification

# Create a sample dataframe
X, y = make_classification(n_samples=1000, n_features=20, n_informative=15, n_redundant=5, random_state=3)

def bootstrap_sample(X, y, n_bootstrap):
    # Generate array of indices for bootstrap sample
    bootstrap_indices = np.random.randint(low=0, high=len(X), size=n_bootstrap)
    return X[bootstrap_indices], y[bootstrap_indices]
```

```python
# Generate 10 bootstrap samples and store their means
means = []
for i in range(1000):
    X_sample, y_sample = bootstrap_sample(X, y, n_bootstrap=X.shape[0])
    means.append(np.mean(X_sample))

import matplotlib.pyplot as plt
# Plot histogram of means
plt.hist(means, bins='auto')
plt.title("Histogram of Bootstrap Sample Means")
plt.xlabel("Mean")
plt.ylabel("Frequency")
plt.show()
```
